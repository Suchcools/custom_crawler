### 爬虫主程序

```python
import requests
from fake_useragent import UserAgent
import json
import pandas as pd
from tqdm import tqdm
import time 

# Create a UserAgent object
user_agent = UserAgent()

# Loop through the desired number of pages
for index, row in tqdm(rawdata.iloc[134:].iterrows()):
        # Define the base URL
        base_url = f'https://apii.mittrchina.com/information/details?id={row.id}'
        # Set the request headers with a random User-Agent
        headers = {
            'User-Agent': user_agent.random,
        }
        try:
            # Send an HTTP GET request
            response = requests.get(base_url, headers=headers)
        except:
            continue
        # Check the response status code
        if response.status_code == 200:
            # Parse the JSON response
            data_json = response.json()
            if 'data' in data_json and data_json['data']!=None:
                # Extract the items and save a json
                with open(f"./detail/{row.id}.json", "w", encoding="utf-8") as json_file:
                    json.dump(data_json['data'], json_file, ensure_ascii=False, indent=4)
            else:
                print(f"No 'data' or 'items' found in the response for ID {row.id}")
        else:
            print(f"Request failed for ID {row.id}, status code: {response.status_code}")
```
